b) ordenada?: defina uma função que recebe uma lista como argumento e devolve o valor booleano
verdadeiro se a lista estiver ordenada e falso caso contrário.

```python
def estaordeandar(lista):
    if len(lista) <= 1:
        return True
    elif lista[0] <= lista[1]:
        return estaordeandar(lista[1:])
    else:
        return False

print(estaordeandar([1, 2, 2, 3, 4, 5]))
```

a) maiorDivisor: defina uma função que recebe um número 
N
, e calcula o maior número de um digito (1 a
9) que divide cada número de entre 
1
 e 
N
. O resultado será um dicionário.

```python
def maiorDivisor(s):
    res = {}
    cond = True
    for i in range(1, s+1):
        cond = True
        for j in range(9, 0, -1):
            if i % j == 0 and cond:
                res[i] = j
                cond = False
    return res

print(maiorDivisor(15))
```

b)  mmc: defina uma função que calcula o mínimo múltiplo comum entre dois números inteiros.

```python
def mmc(a, b):
    maior = a
    if b > a:
        maior = b

    while True:
        if maior % a == 0 and maior % b == 0:
            return maior
        maior += 1

print(mmc(4, 6))
```

c) f2c: Converta um registo de temperaturas Fahrenheit, num novo registo em que a cada temperatura
Fahrenheit associas uma temperatura em Celsius.

```python
def f2C(fahrenheit):
    res = {}
    for chave in fahrenheit:
        celsius = (fahrenheit[chave] - 32) / (1.8)
        res[chave] = round(celsius, 2)
    return res

fahrenheit = {'t1':-30, 't2':-20, 't3':-10, 't4':0} 
print(f2C(fahrenheit)) 
```

d) reduxInt: defina uma função que recebe um número inteiro positivo como argumento e vai adicionar
repetidamente os seu dígitos até obter apenas um dígito que é retornado como resultado.

```python
def reduxInt(s):
    count = 0
    for char in s:
        if char.isdigit():
            count += int(char)
    while count>9:
        newcount = 0
        for char in str(count):
            newcount += int(char)
        count = newcount
    return count

print(reduxInt("998"))
```

```python
def existe_medico(hospital, nome_unidade,nome_medico):
    for unidade in hospital:
        if unidade[0]==nome_unidade:
            for medico in unidade[1]:
                if medico[1]==nome_medico:
                    return True
    return False
print(existe_medico(hospital, "Unidade de Cardiologia", "Dra. Maria"))
```

```python
def ocupar_cama(hospital, nome_unidade, id_cama):
    for unidade in hospital: 
        if unidade[0]==nome_unidade:
            camas=unidade[2]
            for cama in camas:
                if cama[0]==id_cama:
                    cama_index = camas.index(cama)
                    camas[cama_index] = (cama[0], True)
    return hospital
print(ocupar_cama(hospital, "Unidade de Cardiologia", 103))
```

```python
def guardar_hospital(hospital, filename):
    with open(filename, 'w') as f:
        for unidade in hospital:
            f.write(f"Unidade: {unidade[0]}\n")
            f.write("Medicos:\n")
            for medico in unidade[1]:
                f.write(f"  ID: {medico[0]}, Nome: {medico[1]}\n")
            f.write("Camas:\n")
            for cama in unidade[2]:
                status = "Ocupada" if cama[1] else "Disponivel"
                f.write(f"  ID: {cama[0]}, Status: {status}\n")
            f.write("\n")
guardar_hospital(hospital, "hospital.txt")
```

```python
def carregar_hospital(filename):
    hospital = []
    with open(filename, 'r') as f:
        unidade = None
        medicos = []
        camas = []
        for line in f:
            line = line.strip()
            if line.startswith("Unidade:"):
                if unidade is not None:
                    hospital.append((unidade, medicos, camas))
                    medicos = []
                    camas = []
                unidade = line.split("Unidade: ")[1]
            elif line.startswith("Medicos:"):
                continue
            elif line.startswith("Camas:"):
                continue
            elif line.startswith("ID:") and "Nome:" in line:
                parts = line.split(", ")
                id_medico = int(parts[0].split("ID: ")[1])
                nome_medico = parts[1].split("Nome: ")[1]
                medicos.append((id_medico, nome_medico))
            elif line.startswith("ID:") and "Status:" in line:
                parts = line.split(", ")
                id_cama = int(parts[0].split("ID: ")[1])
                status_str = parts[1].split("Status: ")[1]
                ocupado = True if status_str == "Ocupada" else False
                camas.append((id_cama, ocupado))
        if unidade is not None:
            hospital.append((unidade, medicos, camas))
    return hospital

hospital_carregado = carregar_hospital("hospital.txt")
print(hospital_carregado)
```

```python
def camas_ocupadas(hospital):
    ocupadas = 0
    total = 0

    for unidade in hospital:
        camas = unidade[2]
        for cama in camas:
            total += 1
            if cama[1]==True:
                ocupadas += 1

    return ocupadas / total
print(camas_ocupadas(hospital))
```

```python
def unidades_ocupacao_critica(hospital):
    res = []

    for unidade in hospital:
        nome = unidade[0]
        camas = unidade[2]

        total = len(camas)
        ocupadas = 0

        for cama in camas:
            if cama[1]:
                ocupadas += 1

        percentagem = (ocupadas / total) * 100

        if percentagem >= 80:
            res.append((nome, percentagem))

    return res
print(unidades_ocupacao_critica(hospital))
```
